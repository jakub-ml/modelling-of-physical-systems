```python
import matplotlib.pyplot as plt
import numpy as np
import pint
```

The Pint library in Python is used for managing physical quantities with units, making it easy to perform unit-aware computations. It allows users to define, convert, and manipulate values with units (meters, seconds) while automatically checking for consistency in calculations.


This equation describes the concentration c(x,t) of a substance diffusing through a medium over time. It incorporates diffusion ùê∑ and advection with velocity ùë¢, modeling how the substance spreads and moves due to these processes. The equation is commonly used in physics, chemistry, and environmental science.
$$
c(x,t) = \frac{M}{A} \cdot \frac{1}{2 \sqrt{\pi D}} \cdot \frac{x}{t^{3/2}} \cdot \exp \left( -\frac{(ut - x)^2}{4Dt} \right)
$$

```python
# units object
units = pint.UnitRegistry()

# parameters
M = 1.4 * units.kilogram  
A = 0.15 * units.meter * units.meter * units.meter / units.second
D = 3 * units.meter * units.meter / units.second
u = 1.7 * units.meter / units.second

x = np.linspace(60,100,100) * units.meter   
t = np.linspace(0.1,100,100) * units.second
```

```python
# formula
f=(M/A)*(1/(2*np.sqrt(np.pi*D)))*(x/(np.power(t, (3/2))))*np.exp(-(np.power(u*t-x, 2))/(4*D*t))
```

```python
# plot
units.setup_matplotlib()
plt.plot(t,f)
plt.show()
```
